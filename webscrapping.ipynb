```python
import numpy as np
import pandas as pd
```

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
def ind_page_parser(ind):
    url = 'https://www.desmogblog.com/'+ind
    print(url)
    response = requests.get(url)
    l1 = response.text.find('<h2>Resources</h2>') 
    l2 = response.text.find('<h2 class="block__title block-title">Related Posts</h2>')
    soup = BeautifulSoup(response.text[l1:l2], "html.parser")
    res = []
    for link in soup.find_all("a"):
        st = str(format(link.get("href")))
        if ("https://web.archive.org" not in st) & ("http://archive" not in st) & ("https://archive" not in st) & (("http" in st) | ('www' in st)):
            res.append(format(link.get("href")))
    df = pd.DataFrame({'Individual': ind, 'Resources': res})
    return df
```
